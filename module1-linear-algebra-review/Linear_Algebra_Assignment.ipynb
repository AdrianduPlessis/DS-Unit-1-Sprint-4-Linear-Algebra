## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{e} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([1,1,1,8])
h = np.array([3,3,3,3])

g_norm = np.linalg.norm(g)
h_norm = np.linalg.norm(h)

if g_norm > h_norm:
  print("Vector g is longer.")
else:
  print('Vector h is longer.')
```

## 1.9 Show that the following vectors are orthogonal (perpendicular to each other):

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 0 \\ -1  \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}
\end{align}

```
g = np.array([1, 0, -1])
h = np.array([1, np.sqrt(2), 1])

h_dot_g = np.dot(g,h)

if h_dot_g == 0:
  print(f'The two vectors are orthogonal since their dot product is {h_dot_g}.')
else:
  print(f'The two vectors are NOT orthogonal since their dot product is {h_dot_g}, not 0.')
```

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}

3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}

\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
A = [[1,2],
    [3,4],
    [5,6]]

B = [[2,4,6]]

C = [[9,6,3],
    [4,7,11]]

D = [[1,0,0],
    [0,1,0],
    [0,0,1]]

E = [[1,3],
    [5,7]]

matrices_names = ['A', 'B', 'C', 'D', 'E']
matrices = [A, B, C, D, E]
matrices_shapes = []
#Dimentions
print('Dimentions of matrices:')
for i in range(len(matrices)):
  rows = len(matrices[i])
  cols = len(matrices[i][0])
  matrices_shapes.append([rows,cols]) #add to list for reference in later step
  print(f'{matrices_names[i]} has dimentions: ({rows}, {cols})')
  
#Which can be multiplied together
print('\n\nWhich matrices can be multiplied together')
for i in range(len(matrices)):
  for j in range(i+1, len(matrices)): #might be an OBOB i+1
    if matrices_shapes[i][1] == matrices_shapes[j][0]:
      print(f'Matrices {matrices_names[i]} and {matrices_names[j]}, with shapes {matrices_shapes[i]} and {matrices_shapes[j]} respectivly, can be multiplied together and will result in a matrix with the shape [{matrices_shapes[i][0]}, {matrices_shapes[j][1]}]')
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = [[12,11],
     [7,10]]

J = [[0,1,2],
     [7,10,4],
     [3,2,0]]

H_norm = np.linalg.norm(H)
J_norm = np.linalg.norm(J)

print('Norm/Magnatude of H is', H_norm)
print('Norm/Magnatude of J is', J_norm)
```

## 3.3 Find H^{-1} and then find J^{-1}

```
H_inv = np.invert(H)
J_inv = np.invert(J)

print('The inverse of H is:\n', H_inv)
print('The inverse of J is:\n', J_inv)
```

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```
HH_inv = np.matmul(H, H_inv)
G_inv = np.invert(G)
G_invG = np.matmul(G_inv, G)

print('H multiplied by H inverted is:\n', HH_inv)
print('G inverted multiplied by G is:\n',G_invG)
print('Since H==G, this proves that matrix multiplication is not commutative.')

```
